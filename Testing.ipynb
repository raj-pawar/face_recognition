```python
from face_detection_mtcnn import extract_face
from face_recognition import get_embedding, encode_input
from keras.models import load_model
import pickle

model = load_model('model/facenet_keras.h5')
pkl_file = open('celeb_model.sav','rb')
celeb_model = pickle.load(pkl_file)
pkl_file.close()
pkl_file = open('label_encoder.pkl','rb')
le = pickle.load(pkl_file) 
pkl_file.close()
```

```python
def get_prediction(filename):
    face_pixels = extract_face(filename)[0]
    embedding = get_embedding(model,face_pixels)
    embedding = encode_input(embedding.reshape(1,-1))
    pred = le.inverse_transform(celeb_model.predict(embedding))[0]
    print(pred)
```

```python
get_prediction('elton.jpg')
```

```python
get_prediction('ben.jpg')
```
